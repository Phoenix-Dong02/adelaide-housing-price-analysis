# Adelaide Housing Price Analysis (2023–2024)

This project analyses housing price growth across Adelaide suburbs using Python.

## 1. Load Dataset

```python
import pandas as pd
df = pd.read_excel("data/lsg_stats_2024_q4.xlsx")
df.head()
```

## 2. Data Exploration

```python
df.info()
```

### Check Missing Values

```python
df.isnull().sum()
```

## 3. Data Cleaning

```python
df_clean = df.dropna(subset = ["Median Change"])
df_clean.shape
```

## 4. Data Analysis

```python
df_clean.sort_values("Median Change", ascending = False)[["Suburb", "Median 4Q 2023", "Median 4Q 2024", "Median Change"]].head(10)
```

## 5. Data Visualization

```python
df_clean = df_clean.drop_duplicates(subset = "Suburb")
df_clean.shape
```

```python
top10 = df_clean.sort_values("Median Change", ascending = False).head(10)
top10[["Suburb", "Median 4Q 2023", "Median 4Q 2024", "Median Change"]]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.bar(top10["Suburb"], top10["Median Change"])

plt.xticks(rotation=45)

plt.title("Top 10 Adelaide Suburbs by House Price Growth (2023-2024)")

plt.ylabel("Price Growth")

plt.xlabel("Suburb")

plt.show()
```

## Conclusion

The analysis identified the suburbs with the highest housing price growth in Adelaide between 2023 and 2024.

Seacliff and College Park experienced the most significant increases in median house prices.

This suggests strong demand in coastal and inner-city suburbs.
